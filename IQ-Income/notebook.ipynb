##  Introduction:
### What is the intelligence quotient?
The definition states that the intelligence quotient is a measure of intellectual ability. Often IQ is confused with a concrete performance or even education; however, it is the "ability" to perform. In other words, the ability for comprehension, combinations, and learning. (["What is IQ?" Mensa International,](
https://www.mensa.org/iq/what-iq))
A person isn't less intelligent because of a lower level of education. Instead, the one who can achieve the same education with less effort is classified as more intelligent. For example, some intelligence tests detail questions on the names of current politicians. This determines how well the patient can remember a name and its position only by the influence of the media presence.

The intelligence quotient was adapted to a mean value of 100 points. For a standard deviation of 15%, an IQ between 85 and 115 should be noted as normal. Depending on the psychological and physical condition of the patient, the results can also vary by up to 10 points.

Intelligence is not a learning ability but can be consciously increased. Through regular brain training, individual skills are specifically addressed and achieved through a performance increase. Thus, the general thinking-ability can be influenced by each individual within a certain framework. However, this possibility decreases considerably with increasing age. Additionally, part of the intelligence is inherited by the mother and father.

### Criticism and misinterpretation:
The intelligence quotient is not intelligence (["Intelligence Testing Criticisms." Explorable, Explorable.com](
https://explorable.com/intelligence-testing-criticisms)). It is the mathematical result of a test developed by Western Europeans, according to Western European standards in a local environment. It is highly controversial that the same standards can also be applied to other cultures. Emotional intelligence, creativity, practical skills, ambition or diligence are not measured in intelligence tests and are hardly included in any interpolation model.

Many countries in Far East Asia are outliers in this result, because the strict, long and success-oriented school systems there promote precisely those characteristics that are considered optimal for Western societies in intelligence tests. In other countries, there are downward outliers because there are other priorities in the respective cultures.

### The Flynn Effect
It can be observed that in western industrialized countries the IQ increases with the years. This effect, first discovered by James R. Flynn in 1984 (["Flynn Effect." ScienceDirect, Elsevier](https://www.sciencedirect.com/topics/psychology/flynn-effect)), was further researched in the following years and is observed in more studies today. It is assumed that there is an average annual increase of about 0.3 to 0.5 points. The increase is also faster for women than for men. In many countries, however, there is also a reverse Flynn effect, i.e. a decrease in the intelligence quotient.

```python
import numpy as np
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.api import qqplot
df = pd.read_csv("IQ_level.csv")

#Using a for loop to find Outliers in each column of the dataframe
df_mod = df.drop(["rank", "country"], axis=1).columns
outliers  = []
for col in df_mod:
    seventy_fifth = df[f'{col}'].quantile(0.75)
    twenty_fifth = df[f'{col}'].quantile(0.25)
    iqr = seventy_fifth - twenty_fifth
    upper = seventy_fifth + (1.5 * iqr)
    lower = twenty_fifth - (1.5 * iqr)  
    outliers.append((df[f'{col}'] < lower) | (df[f'{col}'] > upper))
   
for col, outlier_values in zip(df_mod, outliers):
    print(f"Outliers in {col}:\n{df[outlier_values]}\n")
    
#Boxplot IQ
sns.boxplot(x=df["IQ"])
plt.title('Boxplot of IQ with Outliers')
plt.xlabel('IQ')
plt.show()

#Boxplot education_expenditure
sns.boxplot(x=df['education_expenditure'])
plt.title('Boxplot of education_expenditure with Outliers')
plt.xlabel('education_expenditure')
plt.show()

#Boxplot avg_income
sns.boxplot(x=df["avg_income"])
plt.title('Boxplot of avg_income with Outliers')
plt.xlabel('avg_income')
plt.show()

#Boxplot avg_temp
sns.boxplot(x=df["avg_temp"])
plt.title('Boxplot of avg_temp with Outliers')
plt.xlabel('avg_temp')
plt.show()

```

### In this phase of our analysis, we shift our focus to the examination of correlations between pairs of variables. Our objective is to explore potential relationships and uncover patterns that may exist within the dataset. We employ linear regression models to predict values for one variable based on another, allowing us to assess the accuracy and effectiveness of these predictive models. This evaluation of predictive performance adds a quantitative dimension to our exploration, enhancing our understanding of the predictive capabilities of the chosen variables.

In this segment, we delve into the intricate relationship between a country's average IQ and its corresponding average income, exploring this connection on a logarithmic scale. The first visualization employs a regression plot with a logarithmic transformation on average income, shedding light on the nature of this relationship.

Subsequently, we venture into predictive modeling, utilizing linear regression to estimate average income based on a nation's average IQ. The model's parameters are examined, and a set of hypothetical data points is generated for varying IQ levels, showcasing the predicted average income.

To assess the accuracy of our model, we present visualizations that provide insights into the distribution of residuals. The 'Residuals vs Fitted' plot illustrates the relationship between the predicted and observed values, while the Q-Q Plot (Quantile-Quantile Plot) assesses the normality of residuals.

These visualizations collectively contribute to a comprehensive understanding of the predictive performance of our model, allowing us to scrutinize the accuracy and reliability of our predictions in the context of the IQ and average income relationship.

```python
import numpy as np
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.api import qqplot
df = pd.read_csv("IQ_level.csv")

#Visualize the relation between IQ and average income (Logaritmic scale)
df['avg_income_log'] = np.log(df['avg_income'])
sns.regplot(x="IQ", y="avg_income_log", data=df, ci=None)
plt.title("Relation between IQ and Income")
plt.xlabel("IQ")
plt.ylabel("Average Income (Logaritmic Scale)")
plt.show()

#Prediction of avgerage income based on the average IQ of a country (linear regression)
iq_vs_income = ols("avg_income_log ~ IQ", data=df).fit()  
print(iq_vs_income.params)
explanatory_data_IQ = pd.DataFrame({"IQ":np.arange(110,170,10)})  
prediction_data_IQ = explanatory_data_IQ.assign(avg_income_log = iq_vs_income.predict(explanatory_data_IQ))

#Visualize predicted data
fig = plt.figure()
sns.regplot(x="IQ", y="avg_income_log", data=df, ci=None)
sns.scatterplot(x="IQ", y="avg_income_log", data=prediction_data_IQ, color="red", marker="s")
plt.title("What if the citizens of a nation had an average IQ level similar to that of Albert Einstein?")
plt.xlabel("Average IQ")
plt.ylabel("Average Income (Logaritmic Scale)")
plt.show()

#Visaulize Accuracy of the plot 
#Residuals vs Fitted
sns.residplot(x="IQ", y="avg_income_log", data=df, lowess = True)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

#Q-Q Plot
qqplot(data=iq_vs_income.resid, fit=True, line="45")
```

### There exists a positive correlation between IQ and average income, with a correlation coefficient of 0.56. However, this correlation, while positive, is not considered exceptionally strong. Now, let's explore whether other variables in our dataset exhibit a more robust relationship with income.

```python
import numpy as np
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.api import qqplot
df = pd.read_csv("IQ_level.csv")
df['log_education_expenditure'] = np.log(df['education_expenditure'])
df['log_avg_income'] = np.log(df['avg_income'])

#Visualize the relation between education expenditure and average income
sns.regplot(x="log_education_expenditure", y="log_avg_income", data=df, ci=None)
plt.title("Relation between Education Expenditure and Average Income")
plt.xlabel("Log Education Expenditure")
plt.ylabel("Log Average Income")
plt.show()

#Prediction of avgerage income based on the average education expenditure of a country (linear regression)
edex_vs_income = ols("log_avg_income ~ log_education_expenditure", data=df).fit()  
print(edex_vs_income.params)
explanatory_data_edex = pd.DataFrame({"log_education_expenditure":np.arange(8,15,1)})  
prediction_data_edex = explanatory_data_edex.assign(log_avg_income = edex_vs_income.predict(explanatory_data_edex))

#Visualize predicted data
fig = plt.figure()
sns.regplot(x="log_education_expenditure", y="log_avg_income", data=df, ci=None)
sns.scatterplot(x="log_education_expenditure", y="log_avg_income", data=prediction_data_edex, color="red", marker="s")
plt.title("What if a country would spend a lot more in education?")
plt.xlabel("Log(Education Expenditure)")
plt.ylabel("log(Average Income)")
plt.show()

#Visaulize Accuracy of the plot 
#Residuals vs Fitted
sns.residplot(x="log_education_expenditure", y="log_avg_income", data=df, lowess = True)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

#Q-Q Plot
qqplot(data=edex_vs_income.resid, fit=True, line="45")
```

### As we could expect, we observe a notably stronger relationship between the amount of money spent on education and the average income of individuals, with a correlation coefficient of 0.86. This suggests a more pronounced and positive association between education expenditure and income levels.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.api import qqplot
df = pd.read_csv("IQ_level.csv")
df['avg_income_log'] = np.log(df['avg_income'])

#Visualize relation between average income and average Temperature
sns.regplot(x="avg_temp", y="avg_income_log", data=df, ci=None)
plt.title("Income and Temperature")
plt.ylabel("Average Income (Logaritmic Scale)")
plt.xlabel("Average Temperature")
plt.show()

#Predict Income based on the temperature of a country
temp_vs_income = ols("avg_income_log ~ avg_temp", data=df).fit()
print(temp_vs_income.params)
explanatory_data_temp = pd.DataFrame({"avg_temp": np.arange(-1, -50, -10)})
prediction_data_temp = explanatory_data_temp.assign(avg_income_log=temp_vs_income.predict(explanatory_data_temp))

#Visualize predicted data
fig = plt.figure()
sns.regplot(x="avg_temp", y="avg_income_log", data=df, ci=None)
sns.scatterplot(x="avg_temp", y="avg_income_log", data = prediction_data_temp, color="red", marker="s")
plt.title("What if we would live in Antartica?")
plt.xlabel("Average Temperature")
plt.ylabel("Average Income (Logaritmic Scale)")
plt.show()

#Visaulize Accuracy of the plot 
#Residuals vs Fitted
sns.residplot(x="avg_temp", y="avg_income", data=df, lowess = True)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

#Q-Q Plot
qqplot(data=temp_vs_income.resid, fit=True, line="45")
```

### While we observe a weak negative correlation (-0.44) between average income and average temperature, it's crucial to note that residing in Antarctica wouldn't necessarily lead to an increase in average income. Numerous other factors contribute to income levels, and these graphs are presented primarily out of curiosity rather than implying a causal relationship between temperature and income.

```python
import numpy as np
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.api import qqplot
df = pd.read_csv("IQ_level.csv")

#Visualize the Relation between IQ and Avarege Temperature
sns.regplot(x="avg_temp", y="IQ", data=df, ci=None)
plt.title("Temperature and IQ")
plt.xlabel("Average Temperature")
plt.ylabel("Average IQ")
plt.show()

#Predict IQ based on the temperature of a country
iq_vs_temp = ols("IQ ~ avg_temp", data=df).fit()
print(iq_vs_temp.params)
explanatory_data_iq_vs_temp = pd.DataFrame({"avg_temp": np.arange(-1, -50, -10)})
prediction_data_iq_vs_temp = explanatory_data_iq_vs_temp.assign(IQ=iq_vs_temp.predict(explanatory_data_iq_vs_temp))

#Visualize predicted data
figure = plt.figure()
sns.regplot(x="avg_temp", y="IQ", data=df, ci=None)
sns.scatterplot(x="avg_temp", y="IQ", data=prediction_data_iq_vs_temp, color="red", marker="s")
plt.title("Cold = Intelligence?")
plt.xlabel("Average Temperature")
plt.ylabel("Average IQ")
plt.show()

#Visaulize Accuracy of the plot 
#Residuals vs Fitted
sns.residplot(x="avg_temp", y="IQ", data=df, lowess = True)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

#Q-Q Plot
qqplot(data=iq_vs_temp.resid, fit=True, line="45")
```

### Upon examining the correlation heatmap, an intriguing observation emerges: there is a notable negative correlation between IQ and Average Temperature (-0.63), seemingly stronger than the correlation between IQ and Average Income (0.56). However, it's important to approach such correlations with caution. Drawing conclusions, such as suggesting an IQ of 140 and above at an average temperature of -40 degrees, would be arbitrary and unfounded. This correlation is highlighted not to imply a direct causal link but rather to underscore the need for careful analysis and interpretation when dealing with diverse datasets.
